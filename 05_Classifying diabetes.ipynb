## diabetes data 분류
* Logistic Classification

```python
import tensorflow as tf
import numpy as np
tf.set_random_seed(777)
```

```python
xy = np.loadtxt('data-03-test-diabetes.csv', delimiter=',', dtype=np.float32)
x_data = xy[:,0:-1]
y_data = xy[:,[-1]]
```

```python
X = tf.placeholder(tf.float32, shape=[None, 8])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W = tf.Variable(tf.random_normal([8, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hypothesis = tf.sigmoid(tf.matmul(X,W)+b)
```

```python
cost = -tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
#if hypothesis>0.5 true, else false
predicted = tf.cast(hypothesis >0.5, dtype = tf.float32)#cast함수는 소수점을 버림
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    feed = {X:x_data, Y:y_data}
    for step in range(10001):
        sess.run(train, feed_dict=feed)
        if step % 200 == 0:
            print(step, sess.run(cost, feed_dict=feed))
            h, p, a = sess.run([hypothesis, predicted, accuracy], feed_dict=feed)
print("\nhypothesis : ",h,"\nPredict : ",p,"\nAccuracy : ",a)
```
